#### Q5: Produce a Fibonacci series of 20 elements using while loop.

```python
#### Q5: Produce a Fibonacci series of 20 elements using while loop.
def fibonacci_generator(n):
    a = 0
    b = 1
    fib_seq = []
    i = 1
    while(i <= n):
        fib_seq.append(a)
        a,b = b,a+b
        i+=1
    return fib_seq

print(fibonacci_generator(20))
```

#### Q6: Produce a Fibonacci series of 15 elements using for loop.

```python
#### Q6: Produce a Fibonacci series of 15 elements using for loop.
def fibonacci_generator(n):
    a = 0
    b = 1
    fib_seq = []
    for i in range (1, n+1):
        fib_seq.append(a)
        a,b = b,a+b
        i+=1
    return fib_seq

print(fibonacci_generator(15))
```

#### Q7: Write a program to check if number is prime or not.

```python
#### Q7: Write a program to check if number is prime or not.
import math


n = 13

is_prime = True
for i in range(2, math.ceil(math.sqrt(n)+1)):
    if(n%i==0):
        is_prime = False
        break

if(is_prime):
    print(f"{n} is prime")
else:
    print(f"{n} is NOT prime")
```

#### Q18: Create a multiplication table (arithmetic table) of 10 using for loop under while loop. Also do vice versa.

```python
#### Q18: Create a multiplication table (arithmetic table) of 10 using for loop under while loop. Also do vice versa.
# using for loop under while loop

n = 10
i = 2
while(i<=n):
    for j in range(1, 11):
        print(f"{i} X {j} = {i*j}")
    print()
    i+=1
```
